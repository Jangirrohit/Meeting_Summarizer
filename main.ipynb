```python

result = graph.invoke(
    {
    "audio_file": "Test_meeting.mp3",
    "transcript_format": "audio"
}, 
config={"configurable": {"thread_id": "user"}}
)
```

```python
result
```

```python
from langgraph.types import Command
if "__interrupt__" in result:
    interrupt_data= result["__interrupt__"][0].value["data"]
    summary = interrupt_data["summary"]
    actions = interrupt_data["action_items"]
    print("\n===========Summary===========")
    print(summary)
    print("\n===========Actions===========")
    for i, action in enumerate(actions, start=1):
        print(f" {i}. {action}")
    decision = input("\nDo you want to execute the actions? (y/n): ").strip().lower()
    if decision == "y":
        print("Executing actions...")
        resume_data= {"approved": True}
    else:
        feedback = input("Please provide feedback on what changes you want to see: ")
        resume_data= {"approved": False, "feedback": feedback}
        print("re-executing.")
    result= graph.invoke(
        Command(resume=resume_data),
        config={"configurable": {"thread_id": "user"}}
    )


    
```
